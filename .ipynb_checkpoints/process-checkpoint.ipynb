```python
import os
import pandas as pd
from sklearn import svm
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
pd.set_option('display.max_rows', 122)
pd.set_option('display.max_columns', 40)
```

```python
## Data collection for push hand pattern
## first window for push hand
first_window = {}
push_df = pd.DataFrame(first_window)
directory = r'1m_data/'
i = 0
for filename in os.listdir(directory):
    if (filename.endswith(".csv")):
        file_path = "1m_data/" + str(filename)
        csv_df = pd.read_csv(file_path)
        
        target_index = 0
        for index in csv_df.index:
            if csv_df.loc[index, 'Time'] > 8.5:
                target_index = index;
                break
        begin_index = int(target_index)
        end_index = int(target_index + 50)
        
        push_hand = csv_df.loc[begin_index:end_index, 'Signal strength (dBm)'].str.replace(' dBm', '')
        push_hand.reset_index(drop = True, inplace = True)
        
        push_hand = push_hand.astype(float)
        push_hand -= push_hand.mean()
        push_df.insert(i, i, push_hand)
        i += 1

push_df.reset_index(drop = True, inplace = True)
push_df.loc[51,:] = 1
push_df = push_df.T


## second window for push hand
second_window = {}
push_df2 = pd.DataFrame(second_window)
directory = r'1m_data/'
i = 0
for filename in os.listdir(directory):
    if (filename.endswith(".csv")):
        file_path = "1m_data/" + str(filename)
        csv_df = pd.read_csv(file_path)
        
        target_index = 0
        for index in csv_df.index:
            if csv_df.loc[index, 'Time'] > 28.5:
                target_index = index;
                break
        begin_index = int(target_index)
        end_index = int(target_index + 50)
        
        push_hand = csv_df.loc[begin_index:end_index, 'Signal strength (dBm)'].str.replace(' dBm', '')
        push_hand.reset_index(drop = True, inplace = True)
        
        push_hand = push_hand.astype(float)
        push_hand -= push_hand.mean()
        push_df2.insert(i, i, push_hand)
        i += 1
        
push_df2.reset_index(drop = True, inplace = True)
push_df2.loc[51,:] = 1
push_df2 = push_df2.T


## third window for push hand
third_window = {}
push_df3 = pd.DataFrame(third_window)
directory = r'1m_data/'
i = 0
for filename in os.listdir(directory):
    if (filename.endswith(".csv")):
        file_path = "1m_data/" + str(filename)
        csv_df = pd.read_csv(file_path)
        
        target_index = 0
        for index in csv_df.index:
            if csv_df.loc[index, 'Time'] > 48.5:
                target_index = index;
                break
        begin_index = int(target_index)
        end_index = int(target_index + 50)
        
        push_hand = csv_df.loc[begin_index:end_index, 'Signal strength (dBm)'].str.replace(' dBm', '')
        push_hand.reset_index(drop = True, inplace = True)
        
        push_hand = push_hand.astype(float)
        push_hand -= push_hand.mean()
        push_df3.insert(i, i, push_hand)
        i += 1
        
push_df3.reset_index(drop = True, inplace = True)
push_df3.loc[51,:] = 1
push_df3 = push_df3.T

push_df = push_df.append(push_df2)
push_df = push_df.append(push_df3)
push_df.reset_index(drop = True, inplace = True)
```

```python
## Data collection for raise hand pattern
## first window for raise hand
first_window = {}
raise_df = pd.DataFrame(first_window)
directory = r'1m_data/'
i = 0
for filename in os.listdir(directory):
    if (filename.endswith(".csv")):
        file_path = "1m_data/" + str(filename)
        csv_df = pd.read_csv(file_path)
        
        target_index = 0
        for index in csv_df.index:
            if csv_df.loc[index, 'Time'] > 12.5:
                target_index = index;
                break
        begin_index = int(target_index)
        end_index = int(target_index + 50)
        
        raise_hand = csv_df.loc[begin_index:end_index, 'Signal strength (dBm)'].str.replace(' dBm', '')
        raise_hand.reset_index(drop = True, inplace = True)
        
        raise_hand = raise_hand.astype(float)
        raise_hand -= raise_hand.mean()
        raise_df.insert(i, i, raise_hand)
        i += 1
        
raise_df.reset_index(drop = True, inplace = True)
raise_df.loc[51,:] = 2
raise_df = raise_df.T


## second window for raise hand
second_window = {}
raise_df2 = pd.DataFrame(second_window)
directory = r'1m_data/'
i = 0
for filename in os.listdir(directory):
    if (filename.endswith(".csv")):
        file_path = "1m_data/" + str(filename)
        csv_df = pd.read_csv(file_path)
        
        target_index = 0
        for index in csv_df.index:
            if csv_df.loc[index, 'Time'] > 32.5:
                target_index = index;
                break
        begin_index = int(target_index)
        end_index = int(target_index + 50)
        
        raise_hand = csv_df.loc[begin_index:end_index, 'Signal strength (dBm)'].str.replace(' dBm', '')
        raise_hand.reset_index(drop = True, inplace = True)
       
        raise_hand = raise_hand.astype(float)
        raise_hand -= raise_hand.mean()
        raise_df2.insert(i, i, raise_hand)
        i += 1
        
raise_df2.reset_index(drop = True, inplace = True)
raise_df2.loc[51,:] = 2
raise_df2 = raise_df2.T

## third window for raise hand
third_window = {}
raise_df3 = pd.DataFrame(third_window)
directory = r'1m_data/'
i = 0
for filename in os.listdir(directory):
    if (filename.endswith(".csv")):
        file_path = "1m_data/" + str(filename)
        csv_df = pd.read_csv(file_path)
        
        target_index = 0
        for index in csv_df.index:
            if csv_df.loc[index, 'Time'] > 52.5:
                target_index = index;
                break
        begin_index = int(target_index)
        end_index = int(target_index + 50)
        
        raise_hand = csv_df.loc[begin_index:end_index, 'Signal strength (dBm)'].str.replace(' dBm', '')
        raise_hand.reset_index(drop = True, inplace = True)
       
        raise_hand = raise_hand.astype(float)
        raise_hand -= raise_hand.mean()
        raise_df3.insert(i, i, raise_hand)
        i += 1
        
raise_df3.reset_index(drop = True, inplace = True)
raise_df3.loc[51,:] = 2
raise_df3 = raise_df3.T

raise_df = raise_df.append(raise_df2)
raise_df = raise_df.append(raise_df3)
raise_df.reset_index(drop = True, inplace = True)
```

```python
## Data collection for hold hand pattern
## first window for hold hand
first_window = {}
hold_df = pd.DataFrame(first_window)
directory = r'1m_data/'
i = 0
for filename in os.listdir(directory):
    if (filename.endswith(".csv")):
        file_path = "1m_data/" + str(filename)
        csv_df = pd.read_csv(file_path)
        
        target_index = 0
        for index in csv_df.index:
            if csv_df.loc[index, 'Time'] > 10:
                target_index = index;
                break
        begin_index = int(target_index)
        end_index = int(target_index + 50)
        
        hold_hand = csv_df.loc[begin_index:end_index, 'Signal strength (dBm)'].str.replace(' dBm', '')
        hold_hand.reset_index(drop = True, inplace = True)
        
        hold_hand = hold_hand.astype(float)
        hold_hand -= hold_hand.mean()
        hold_df.insert(i, i, hold_hand)
        i += 1
        
hold_df.reset_index(drop = True, inplace = True)
hold_df.loc[51,:] = 3
hold_df = hold_df.T

## second window for hold hand
second_window = {}
hold_df2 = pd.DataFrame(second_window)
directory = r'1m_data/'
i = 0
for filename in os.listdir(directory):
    if (filename.endswith(".csv")):
        file_path = "1m_data/" + str(filename)
        csv_df = pd.read_csv(file_path)
        
        target_index = 0
        for index in csv_df.index:
            if csv_df.loc[index, 'Time'] > 30:
                target_index = index;
                break
        begin_index = int(target_index)
        end_index = int(target_index + 50)
        
        hold_hand = csv_df.loc[begin_index:end_index, 'Signal strength (dBm)'].str.replace(' dBm', '')
        hold_hand.reset_index(drop = True, inplace = True)

        hold_hand = hold_hand.astype(float)
        hold_hand -= hold_hand.mean()
        hold_df2.insert(i, i, hold_hand)
        i += 1
        
hold_df2.reset_index(drop = True, inplace = True)
hold_df2.loc[51,:] = 3
hold_df2 = hold_df2.T

## third window for hold hand
third_window = {}
hold_df3 = pd.DataFrame(third_window)
directory = r'1m_data/'
i = 0
for filename in os.listdir(directory):
    if (filename.endswith(".csv")):
        file_path = "1m_data/" + str(filename)
        csv_df = pd.read_csv(file_path)
        
        target_index = 0
        for index in csv_df.index:
            if csv_df.loc[index, 'Time'] > 50:
                target_index = index;
                break
        begin_index = int(target_index)
        end_index = int(target_index + 50)
        
        hold_hand = csv_df.loc[begin_index:end_index, 'Signal strength (dBm)'].str.replace(' dBm', '')
        hold_hand.reset_index(drop = True, inplace = True)

        hold_hand = hold_hand.astype(float)
        hold_hand -= hold_hand.mean()
        hold_df3.insert(i, i, hold_hand)
        i += 1
        
hold_df3.reset_index(drop = True, inplace = True)
hold_df3.loc[51,:] = 3
hold_df3 = hold_df3.T

hold_df = hold_df.append(hold_df2)
hold_df = hold_df.append(hold_df3)
hold_df.reset_index(drop = True, inplace = True)
```

```python
df = push_df.append(raise_df)
df = df.append(hold_df)
df.reset_index(drop = True, inplace = True)
```

```python
X = df.drop(columns=[51])
Y = df[51]
Y = Y.astype(int)
Y = Y.astype(str)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2)
model = DecisionTreeClassifier()
model.fit(X_train, Y_train)
predictions = model.predict(X_test)

total_score = 0
for i in range(10000):
    score = accuracy_score(Y_test, predictions)
    total_score += score;
average_score = total_score / 10000;
average_score
```

```python
clf = svm.SVC()
clf.fit(X_train, Y_train)
predictions = clf.predict(X_test)

total_score = 0
for i in range(10000):
    score = accuracy_score(Y_test, predictions)
    total_score += score;
average_score = total_score / 10000;
average_score
```
